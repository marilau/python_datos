### Cargar la información de alumnos a una base de datos, y realizar el cruce con establecimientos

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:1324pg@127.0.0.1/mod3')
conn = engine.connect()
```

```python
result = conn.execute('select * from alumnos')
alumnos = pd.DataFrame(result.fetchall())
```

```python
alumnos.head()
```

```python
alumnos.columns = result.keys()
alumnos.head()
```

```python
mae_chaco = mae[mae.jurisdiccion == 'Chaco']
mae_chaco.head()
```
