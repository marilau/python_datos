```python
from sqlalchemy import create_engine
```

```python
import pandas as pd
```

```python
engine = create_engine('postgresql://postgres:1324pg@127.0.0.1/mod3')
conn = engine.connect()
```

```python
result = conn.execute('select * from agrupar')
df = pd.DataFrame(result.fetchall())
```

```python
df
```

```python
df.columns = result.keys()
```

```python
df
```

```python
conn.close()
```

```python
df.groupby('gender')
```

```python
df.groupby('gender').mean()
```

```python
df.groupby('gender').mean().round(2)
```

```python
df[['gender','salary']].groupby('gender').mean().round(2)
```

```python
df.groupby('gender').agg(avg_salary=pd.NamedAgg('salary', 'mean')).round(2)
```

```python
df.groupby('gender').agg(avg_salary=('salary', 'mean')).round(2)
```

```python
agrupado = df.groupby('gender').agg(
    median_salary = ('salary', 'median'),
    avg_height = ('height', 'mean')
)
```
