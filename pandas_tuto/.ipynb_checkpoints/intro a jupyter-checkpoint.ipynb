### Filtrado con función query

```python
locs.query('CATEGORIA=="Localidad simple" and PROVINCIA_NOMBRE=="Chaco"')
```

```python
locs.query('CATEGORIA=="Localidad simple" or PROVINCIA_NOMBRE=="Chaco"')
```

```python
provs = ['Chaco', 'Misiones', 'Santa Fe']
locs.query('PROVINCIA_NOMBRE in @provs')
```
